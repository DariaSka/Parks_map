```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
file_path = 'park_info_file.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
# Create a map centered on a specific location (latitude, longitude)
map_center = [56.130366, -106.346771]
map_zoom = 4  # Adjust the zoom level as needed
my_map = folium.Map(location=map_center, zoom_start=map_zoom)
```

```python
points_coordinates = []
```

```python
for i in range(df.shape[0]):
  if (df.Ispark[i] == True):
    points_coordinates.append([df.Lat[i],df.Lon[i],df.Full_name[i],df.freq_count2[i],df.URL[i]])
```

```python
points_coordinates[0]
```

```python
def get_marker_color(weight):
    if weight < 4:
        return 'royalblue'
    elif weight < 7:
        return 'seagreen'
    elif weight < 12:
        return 'gold'
    elif weight < 25:
        return 'darkorange'
    else:
        return 'firebrick'
```

```python
def get_popup(name, weight, url):
  short_name = name.replace(" Of Canada","").replace(" of Canada","")
  popup = f'''
    <h4>{short_name}</h4>
    <p>Number of mentions: {weight}</p>
    <a href={url} target="_blank">See on Google Maps</a>
  '''
  #popup = name.replace(" Of Canada","").replace(" of Canada","")
  #popup = popup + ", " + str(coord[3]) + "\n" + url
  return popup
```

```python
for coord in points_coordinates:
     latlon = [coord[0],coord[1]]
     marker = folium.CircleMarker(
        location=latlon,
        radius= min(4 + int(coord[3]),20),  # Adjust the marker size based on weight.
        color=get_marker_color(coord[3]),  # You can also use other colors or set a color range based on weight.
        fill=True,
        fill_color=get_marker_color(coord[3]),
        fill_opacity=0.8,
        popup= folium.Popup(get_popup(coord[2], coord[3], coord[4]), max_width=300)  ,  # Display the name and weight in the popup.
        #icon = None
     )

     dicon = folium.DivIcon(
        html=f'<div style="font-size: 6px; font-weight: bold;">{coord[3]}</div>',
        icon_size=(30, 30)  # Adjust the icon size as needed
     )
     marker.icon = dicon  # Use create_icon to generate the icon object

     # Add the marker to the map.
     marker.add_to(my_map)
```
